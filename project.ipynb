```python
# extract the features words 
from sklearn.feature_extraction.text import TfidfVectorizer

vec=TfidfVectorizer(max_features=4000,stop_words='english')

vector=vec.fit_transform(df['Tags']).toarray()

print(vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sim=cosine_similarity(vector)

print(sim)
```

```python
# recommend positions
def file (file):
    with plu.open(file) as pdf:
        page=pdf.pages[0]
        text=page.extract_text()
        par=text.replace("\n"," ")
    rvec=vec.transform([par]).toarray()
    sim=cosine_similarity(rvec,vector)[0]
    df['sim']=sim
    df_new=df.sort_values(by='sim',ascending=False).iloc[:50]
    lis=df_new['Position'].unique().tolist()
    print(f" 1 {lis[0]} \n 2 {lis[1]} \n 3 {lis[2]}")
```

```python
# compare the resume
files=["Resume 1.pdf","it_resume.pdf","resume tronics.pdf"]
for file in files:
    with plu.open(file) as pdf:
        page=pdf.pages[0]
        text=page.extract_text()
        par=text.replace("\n"," ")
        print(file)
        rvec=vec.transform([par]).toarray()
        rsim=cosine_similarity(rvec,vector)[0]
        df[f"{file}"]=rsim
```
